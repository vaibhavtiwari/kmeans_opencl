```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.subplots(figsize = (30,10))
plot1 = plt.subplot2grid((1, 2), (0, 0))
plot2 = plt.subplot2grid((1, 2), (0, 1))

# Path for CPU output csv file
dfCpu = pd.read_csv("outputCPU.csv")
groups = dfCpu.groupby('c')
for name, group in groups:
    plot1.plot(group.x, group.y, marker='o',linestyle='', label = name)
    plot1.set_title("CPU Clustering Output")
    
# Path for GPU output csv file  
dfGpu = pd.read_csv("outputGPU.csv")
groupsGpu = dfGpu.groupby('c')
for name, group in groupsGpu:
    plot2.plot(group.x, group.y, marker='o',linestyle='', label = name)
    plot2.set_title("GPU Clustering Output")
    
plot1.legend()
plot2.legend()
```
